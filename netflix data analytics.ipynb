```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('/content/netflix_titles.csv.zip')
df.head()
```

```python
df.drop_duplicates(inplace=True)
df.isnull().sum()
```

```python
df['type'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Content Type Distribution')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()
```

```python
genres = df['listed_in'].str.split(', ').explode()
genres.value_counts().head(10).plot(kind='barh', color='salmon')
plt.title('Top 10 Genres')
plt.xlabel('Count')
plt.ylabel('Genre')
plt.show()
```

```python
df['release_year'].value_counts().sort_index().plot(kind='line', figsize=(12,6), color='green')
plt.title('Content Release Trend Over Years')
plt.xlabel('Year')
plt.ylabel('Number of Titles')
plt.grid(True)
plt.show()
```

```python
df['country'].value_counts().head(10).plot(kind='bar', color='purple')
plt.title('Top 10 Countries by Content')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45)
plt.show()
```

```python
df['rating'].value_counts().plot(kind='pie', autopct='%1.1f%%', figsize=(8,8), startangle=90)
plt.title('Content Rating Distribution')
plt.ylabel('')
plt.show()
```

##  Insights from the Data
- Netflix has more Movies than TV Shows overall.
- The most popular genres include Dramas and Comedies.
- The highest number of releases happened in 2018 and 2019.
- Most Netflix content comes from the United States.
- TV-MA is the most common rating on the platform.
